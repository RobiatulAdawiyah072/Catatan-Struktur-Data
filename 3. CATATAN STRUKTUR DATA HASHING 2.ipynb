TABEL HASHING adalah kumpulan item yang disimpan sedemikian rupa untuk memudahkan menemukannya nanti. Setiap posisi tabel hash, sering disebut slot, dapat menyimpan item dan diberi nama oleh nilai integer mulai dari 0. Sebagai contoh, kita akan memiliki slot bernama 0, slot bernama 1, slot bernama 2, dan seterusnya di. Awalnya, tabel hash tidak berisi item sehingga setiap slot kosong. Kita dapat mengimplementasikan tabel hash dengan menggunakan daftar dengan setiap elemen diinisialisasi dengan nilai None. HASHING terbagi menjadi 3 yaitu LINEAR PROBBING, QUADRATIC PROBBING, dan CHAINING.


1. LINEAR PROBBING
    LINEAR PROBBING adalah cara penyelesaian collision (terjadi tumbukan) pada hashing dengan memindahkan angka yang collision dengan collision=collision+1 % panjangList,  sehingga ketika ada collision pada index 0, maka akan di tempatkan pada 0 + 1 % panjangList maka, akan ditempatkan pada index 1. namun saat pada index 1 terdapat item didalamnya maka akan bergeser lagi pada index 1 + 1 % panjangList (index menjadi 2) begitu seterusnya.
    
2. QUADRATIC PROBBING
    QUADRATIC PROBBING adalah adalah cara penyelesaian collision pada hashing dengan memindahkan angka yang collision dengan collision=collision + (i^2) % panjang_index (i bermula dari angka 1 dengan increment +1) sehingga kita akan tempatkan item yang bertabrakan pada index +(1^2,2^2,dst) hingga menemukan posisi yang kosong.
    
3. CHAINING 
    CHAINING adalah cara penyelesain collision pada hashing dengan menyatukan angka yang collision di dalam satu tabel hashing.

```python
#proses hashing
table = int(input("panjang tabel hash : "))
table = [None] * table
print(table)
def hash(x,table):
    return x % len(table)

def insert(table,key,value):
    index = hash(key,table)
    if table[index] == None:
        table[index] = value
    else:
        collusion = index
        found = False
        ind = collusion + 1
        if ind > len(table) - 1:
                ind = 0
        while (ind <= len(table)-1) and not found:
            if table[ind] == None:
                found = True
                table[ind] = value
            ind = ind + 1

a = "y"
while a == "y" :
    jumlah = int(input("jumlah angka yang ingin dimasukkan : "))
    if jumlah > len(table):
        print("jumlah angka yang dimasukkan terlalu panjang")
        print("masukkan lagi")
    else :
        a = "t"

#untuk memasukkan angka
angka = []
for p in range (jumlah):
    masukan = int(input("Masukan angka = "))
    insert(table,masukan,masukan)
    p = p + 1
    angka.append(masukan)
print(angka,
      "<<<<<<<<<<<<< tabel sebelum hashing")
print(table,
      "<<<<<<<<<<<<< tabel setelah Hashing")
```

```python
#CONTOH CODE QUADRATIC PROBBING

table = int(input("panjang tabel hash : "))
table = [None] * table
print(table)

def hash(x,table):
    return x % len(table)

def insert(table,key,value):
    index = hash(key,table)
    if table[index] == None:
        table[index] = value
    else:
        collusion = index
        found = False
        i = 1
        while i > 0 and not found :
            ind = collusion + i**2
            ind = ind % len(table)
            if table[ind] == None :
                found = True
                table[ind] = value
            i = i + 1

#untuk menginputkan jumlah angka
a = "y"
while a == "y" :
    jumlah = int(input("jumlah angka yang ingin dimasukkan : "))
    if jumlah > len(table):
        print("jumlah angka yang dimasukkan terlalu panjang")
        print("masukkan lagi")
    else :
        a = "t"

#untuk memasukkan angka
angka = []
for p in range (jumlah):
    masukan = int(input("Masukan angka = "))
    insert(table,masukan,masukan)
    p = p + 1
    angka.append(masukan)
print(angka,
      "<<<<<<<<<<<<< tabel sebelum hashing")
print(table,
      "<<<<<<<<<<<<<Tabel setelah hashing")
```

```python
#CONTOH CODE CHAINING

table = int(input("panjang tabel hash : "))
table = [[] for i in range(table)]
print(table)

def hash(x,table):
    return x % len(table)

def insert(table,key,value):
    table[hash(key,table)].append(value)
    
#untuk menginputkan jumlah angka
a = "y"
while a == "y" :
    jumlah = int(input("jumlah angka yang ingin dimasukkan : "))
    if jumlah > len(table):
        print("jumlah angka yg ingin dimasukkan terlalu panjang")
        print("masukkan lagi")
    else :
        a = "t"

#untuk memasukkan angka
angka = []
for p in range (jumlah):
    masukan = int(input("Masukan angka = "))
    insert(table,masukan,masukan)
    p = p + 1
    angka.append(masukan)
print(angka,
      "<<<<<<<<<<<<< tabel sebelum hashing")
print(table,
      "<<<<<<<<<<<<< tabel setelah hasing")
```
